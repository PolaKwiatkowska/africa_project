```python
selection = [ 'Nigeria', 'Egypt', 'South Africa', 'Kenya', 'Algeria', 'Morocco', 'Angola', 'Ghana', 'Cameroon', "Cote d''Ivoire"]
sex = ['Women', 'Men']

#Create dataframe for country from selection list only
sel = pd.DataFrame(selection)
dataset = pd.read_csv("/Users/pola/Documents/Africa Project/bmi.csv", delimiter = ';')
#dataset.head
#dataset_selected = dataset[dataset['country'].isin(selection)]

#dataset_selected
    
```

```python


source_men_A = dataset.loc[((dataset["sex"] == "Men") & (dataset["country"] == "Nigeria") & (dataset["year"] >= 2000) & (dataset["year"] <= 2014))]
source_women_A = dataset.loc[((dataset["sex"] == "Women") & (dataset["country"] == "Nigeria") & (dataset["year"] >= 2000) & (dataset["year"] <= 2014))]

bmi_men = source_men_A["bmi_mean"]
bmi_women = source_women_A["bmi_mean"]
year = source_men_A["year"]


# create a new plot with a title and axis labels
p1 = figure(x_axis_label='year', y_axis_label='bmi')

# add a line renderer with legend and line thickness to the plot
p1.line(x = year, y = bmi_men, line_color="red", legend_label="Men", line_width=2)
p1.line(x = year, y = bmi_women,line_color="green",legend_label="Women", line_width=2)
p1.legend.location = "bottom_right"
p1.add_layout(Title(text=" ", text_font_style="italic", text_font_size = "7pt"), 'above')
p1.add_layout(Title(text="BMI", text_font_size="9pt"), 'above')
p1.xaxis.major_label_orientation = "horizontal"


hover1 = HoverTool()
hover1.tooltips = [("BMI", "$y{0.00}")]
hover1.mode = 'vline'
p1.add_tools(hover1)

# show the results
#show(p)
```

```python
hypertension = pd.read_csv("/Users/pola/Documents/Africa Project/hypertension.csv", delimiter = ';')

source_men_B = hypertension.loc[((hypertension["sex"] == "Men") & (hypertension["country"] == "Nigeria") & (hypertension["year"] >= 2000) & (hypertension["year"] <= 2014))]
source_women_B = hypertension.loc[((hypertension["sex"] == "Women") & (hypertension["country"] == "Nigeria") & (hypertension["year"] >= 2000) & (hypertension["year"] <= 2014))]

hypertension_men = source_men_B["hypertension_prevalence"]
hypertension_women = source_women_B["hypertension_prevalence"]
year = source_men_B["year"]


# create a new plot with a title and axis labels
p2 = figure(x_axis_label='year', y_axis_label='hypertension')

# add a line renderer with legend and line thickness to the plot
p2.line(x = year, y = hypertension_men, line_color="red", legend_label="Men", line_width=2)
p2.line(x = year, y = hypertension_women, line_color="green", legend_label="Women", line_width=2)
p2.legend.location = "bottom_right"

p2.add_layout(Title(text=" ", text_font_style="italic", text_font_size = "7pt"), 'above')
p2.add_layout(Title(text="Hypertension Prevalence", text_font_size="9pt"), 'above')
p2.xaxis.major_label_orientation = "horizontal"


hover2 = HoverTool()
hover2.tooltips = [("Hypertension", "$y")]
hover2.mode = 'vline'
p2.add_tools(hover2)


# show the results
#show(p)

diabetes = pd.read_csv("/Users/pola/Documents/Africa Project/diabetes.csv", delimiter = ';')

source_men_C = diabetes.loc[((diabetes["sex"] == "Men") & (diabetes["country"] == "Nigeria") & (diabetes["year"] >= 2000) & (diabetes["year"] <= 2014))]
source_women_C = diabetes.loc[((diabetes["sex"] == "Women") & (diabetes["country"] == "Nigeria") & (diabetes["year"] >= 2000) & (diabetes["year"] <= 2014))]

diabetes_men = source_men_C["diebetes_prevalence"]
diabetes_women = source_women_C["diebetes_prevalence"]
year = source_men_C["year"]


# create a new plot with a title and axis labels
p3 = figure(x_axis_label='year', y_axis_label='diabetes')

# add a line renderer with legend and line thickness to the plot
p3.line(x = year, y = diabetes_men, line_color="red", legend_label="Men", line_width=2)
p3.line(x = year, y = diabetes_women, line_color="green", legend_label="Women", line_width=2)
p3.legend.location = "bottom_right"

p3.add_layout(Title(text=" ", text_font_style="italic", text_font_size = "7pt"), 'above')
p3.add_layout(Title(text="Diabetes prevalence", text_font_size="9pt"), 'above')
p3.xaxis.major_label_orientation = "horizontal"


hover3 = HoverTool()
hover3.tooltips = [("Diabetes", "$y")]
hover3.mode = 'vline'
p3.add_tools(hover3)

# adding private health expenditure
prv_health_exp = pd.read_csv("/Users/pola/Documents/Africa Project/priv_health_exp.csv", delimiter = ';')

prive_exp = prv_health_exp.loc[(prv_health_exp["country"] == "Nigeria") & (prv_health_exp["year"] >= 2000) & (prv_health_exp["year"] <= 2014)]
indicator_D = prive_exp["value"]
year_D = prive_exp["year"]

p4 = figure(x_axis_label='year', y_axis_label='% of current health expenditure')
p4.line(x = year_D, y = indicator_D, line_color="red", line_width=2)

p4.add_layout(Title(text="(% of current health expenditure)", text_font_style="italic", text_font_size = "7pt"), 'above')
p4.add_layout(Title(text="Domestic private health expenditure", text_font_size="9pt"), 'above')
p4.xaxis.major_label_orientation = "horizontal"


hover4 = HoverTool()
hover4.tooltips = [("Value", "$y{0.00}")]
hover4.mode = 'vline'
p4.add_tools(hover4)


# adding GDP
gdp = pd.read_csv("/Users/pola/Documents/Africa Project/gdp_per_cap.csv", delimiter = ';')

gdp = gdp.loc[(gdp["country"] == "Nigeria") & (gdp["year"] >= 2000) & (gdp["year"] <= 2014)]
indicator_E = gdp["value"]
year_E = gdp["year"]

p5 = figure(x_axis_label='year', y_axis_label='gdp per capita')
p5.line(x = year_E, y = indicator_E, line_color="red", line_width=2)

p5.add_layout(Title(text="(current US$)", text_font_style="italic", text_font_size = "7pt"), 'above')
p5.add_layout(Title(text="GDP per capita", text_font_size="9pt"), 'above')
p5.xaxis.major_label_orientation = "horizontal"


hover5 = HoverTool()
hover5.tooltips = [("Value", "$y{0.00}")]
hover5.mode = 'vline'
p5.add_tools(hover5)


# adding population
pop = pd.read_csv("/Users/pola/Documents/Africa Project/pop.csv", delimiter = ';')

pop = pop.loc[(pop["country"] == "Nigeria") & (pop["year"] >= 2000) & (pop["year"] <= 2014)]
indicator_F = pop["value"]
year_F = pop["year"]

p6 = figure(x_axis_label='year', y_axis_label='population')
p6.line(x = year_F, y = indicator_F, line_color="red", line_width=2)

p6.add_layout(Title(text=" ", text_font_style="italic", text_font_size = "7pt"), 'above')
p6.add_layout(Title(text="Population", text_font_size="9pt"), 'above')
p6.xaxis.major_label_orientation = "horizontal"
p6.yaxis.formatter=NumeralTickFormatter(format="0.00 a")

hover6 = HoverTool()
hover6.tooltips = [("Value", "$y{0.00 a}")]
hover6.mode = 'vline'
p6.add_tools(hover6)

# adding goverment health expenditure
gov_exp = pd.read_csv("/Users/pola/Documents/Africa Project/gov_exp.csv", delimiter = ';')

gov_exp = gov_exp.loc[(gov_exp["country"] == "Nigeria") & (gov_exp["year"] >= 2000) & (gov_exp["year"] <= 2014)]
indicator_G = gov_exp["value"]
year_G = gov_exp["year"]

p7 = figure(x_axis_label='year', y_axis_label='Goverment Health Expenditure')
p7.line(x = year_G, y = indicator_G, line_color="red", line_width=2)

p7.add_layout(Title(text="PPP (current international $)", text_font_style="italic", text_font_size = "7pt"), 'above')
p7.add_layout(Title(text="Domestic general government health expenditure per capita", text_font_size="9pt"), 'above')
p7.xaxis.major_label_orientation = "horizontal"


hover7 = HoverTool()
hover7.tooltips = [("Value", "$y{0.00}")]
hover7.mode = 'vline'
p7.add_tools(hover7)

# adding foreign direct investment
for_inv = pd.read_csv("/Users/pola/Documents/Africa Project/for_direct_inv.csv", delimiter = ';')

for_inv = for_inv.loc[(for_inv["country"] == "Nigeria") & (for_inv["year"] >= 2000) & (for_inv["year"] <= 2014)]
indicator_H = for_inv["value"]
year_H = for_inv["year"]

p8 = figure(x_axis_label='year', y_axis_label='Goverment Health Expenditure')
p8.line(x = year_H, y = indicator_H, line_color="red", line_width=2)

p8.add_layout(Title(text="net inflows (BoP, current US$)", text_font_style="italic", text_font_size = "7pt"), 'above')
p8.add_layout(Title(text="Foreign direct investment", text_font_size="9pt"), 'above')
p8.xaxis.major_label_orientation = "horizontal"
p8.yaxis.formatter=NumeralTickFormatter(format="0.00 a")


hover8 = HoverTool()
hover8.tooltips = [("Value", "$y{0.00 a}")]
hover8.mode = 'vline'
p8.add_tools(hover8)



grid = gridplot([p1, p2, p3, p4, p5, p6, p7, p8], ncols = 4,sizing_mode="scale_both")
show(grid)
#show(row(children=[p1, p2, p3, p4, p5], sizing_mode="scale_width")
```
