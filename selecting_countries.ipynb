```python
import pandas as pd

df = pd.read_csv('/Users/pola/Documents/Africa Project/Africa_City_poplulation.csv', delimiter = ';')
df
```

```python
gdp_df = pd.read_csv('/Users/pola/Documents/Africa Project/gdp.csv', delimiter = ';',decimal=",")
gdp_df.columns = ['country', 'GDP_country_2019']
gdp_df
```

```python
african_countries = df.country.unique()
print(african_countries)
```

```python
africa_df = pd.DataFrame(african_countries, columns=['country'])

Left_join = pd.merge(africa_df, 
                     gdp_df, 
                     on ='country', 
                     how ='left')
Left_join
```

```python
pop_df = pd.read_csv('/Users/pola/Documents/Africa Project/Population_by_country_2019.csv', delimiter = ';')
pop_df.columns = ['country', 'population_country_2019']
pop_df
```

```python
final_df = pd.merge(Left_join, 
                     pop_df, 
                     on ='country', 
                     how ='left')
final_df
```

```python
top25_gdp = final_df.nlargest(25, ['GDP_country_2019', 'population_country_2019'])
selection_10_countires = top25_gdp.sort_values(by = 'population_country_2019',ascending=False).head(10)
print(selection_10_countires)
```

```python
# Selection on citites from 10 selected countries. Cities criterion --> city population greater than 900k 

cities_selection = pd.merge(selection_10_countires, 
                     df, 
                     on ='country', 
                     how ='left')

#cities_selection['population_2021'] = cities_selection['population_2021'].str.replace(" ", "")
cities_selection['population_2021'] = pd.to_numeric(cities_selection['population_2021'], downcast='float')

cities_selection_900k = cities_selection['population_2021'] >= 900000

filtered_df = cities_selection[cities_selection_900k]
print(filtered_df)


```
